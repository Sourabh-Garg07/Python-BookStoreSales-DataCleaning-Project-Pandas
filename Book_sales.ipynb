```python
## IMPORTING NECESSARY LIBRARIES

import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
## IMPORTING DATASET

books = pd.read_csv(r"C:\Users\hinaj\OneDrive\Documents\Project datasets\Book sales.csv")
books.head()
```

```python
## CREATING COPY OF THE DATASET 

df = books.copy()
df.head()
```

```python
## EXPLORING DATASET

df.info()

## Most of the datset has object datatype
## We need to change datatype for PurchaseDate From object to datetime 
```

```python
df.shape

## Dataset has 2000 rows and 8 columns 
```

```python
df.describe()

## Statistical description of data with numbers 
```

```python
## STARTING WITH DATA CLEANING STEPS 

## 1. All the columns are relevant in the dataset except Email which has 185 null values and other irrelevant values.
##    As we do not need email in our analysis we will remove that column

df.drop(columns=["Email"], inplace=True)
df
```

```python
## CHECKING NULL VALUES 

df.isnull().sum()

## Phone number and Address has some null values but the important columns like Name, purchase amount,
## book title does not contain any null values so we will not delete any rows with null values.
```

```python
## CHECKING DUPLICATE RECORDS 

df.duplicated().sum()

## Data does not contains any duplicated records however we will also check by removing customer id column
```

```python
df.duplicated(subset=['Name', 'PhoneNumber', 'Address', 'PurchaseAmount', 'PurchaseDate', 'BookTitle']).sum()

## Data does not contain any duplicated records 
```

```python
## CHECKING COLUMN NAME 

df['Name']
```

```python
## CHECKING FOR INCONSISTENCIES 

df['Name'].str.replace('[A-Za-z]', '', regex=True).unique()

## We do not have any odd charcters in the name but the result shows [' '] which defines that we have some 
## unwanted spaces in the column which we will remove 
```

```python
df['Name'] = df['Name'].str.strip(" ")
```

```python
## CONVERTING NAMES INTO UPPER 

df['Name'] = df['Name'].str.upper()
```

```python
## LOOKING AT BOOK TITLE COLUMN 

df['BookTitle'].head(20)

## Every title has a (.) in the end we will strip that out 

```

```python
## CHECKING OTHER INCONSISTENICIES 

df['BookTitle'].str.replace('[A-Za-z]', '', regex=True).unique()

## We can see that the column contains irregular spaces and (.) in them 
```

```python
## REMOVING INCONSISTENCIES 

df['BookTitle'] = df['BookTitle'].str.strip('. ')
```

```python
## CONEVRTING BOOK TITLE INTO UPPER CASE LETTERS

df['BookTitle'] = df['BookTitle'].str.upper()
```

```python
## CONVERTING DATATYPE OF PURCHASE DATE INTO DATETIME 

df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'], format='mixed', errors='coerce')
df.info()
```

```python
df
```

```python
## LOOKING AT ADDRESS COLUMN

df['Address'].head(20)

## We will split the adress columns into 3 columns - "Street address" , "city/locality", "state"
```

```python
df[['StreetAddress', 'City/Locality', 'State']] = df['Address'].str.split(',', expand=True)
df
```

```python
## CHECKING PHONE NUMBER COLUMN 

df['PhoneNumber']

## The format of phone number is inconsistent throughout some contains extension number while some does not 
## so to make it consistent we will remove the extensions throughout as it is not very useful for our analysis
## and we will convert the phone number in the format : XXX-XXX-XXXX 
```

```python
## CHECKING INCONSISTENICIES 

df['PhoneNumber'].str.replace('[0-9]', '', regex=True).unique()
```

```python
## REMOVING EXTENSIONS 

df['PhoneNumber'] = df['PhoneNumber'].str.replace(r'x\d+', '', regex=True)
```

```python
## REMOVING INCOSNSITENCIES 

df['PhoneNumber'] = df['PhoneNumber'].str.replace('.', '')
df['PhoneNumber'] = df['PhoneNumber'].str.replace('-', '')
df['PhoneNumber'] = df['PhoneNumber'].str.replace('(', '')
df['PhoneNumber'] = df['PhoneNumber'].str.replace(')', '')
df['PhoneNumber'] = df['PhoneNumber'].str.replace('+', '')

df['PhoneNumber']
```

```python
## Some phone numbers contains code 001 in the beginning which we need to extract out 

df['PhoneNumber'] = df['PhoneNumber'].str.replace(r'^001', '', regex=True)
df['PhoneNumber'].head()
```

```python
df['PhoneNumber'].str.replace('[0-9]', '', regex=True).unique()

## Column has been removed of all inconsistenicies
```

```python
## CHANGING FORMAT OF THE PHONE NUMBER 

df['PhoneNumber'] = df['PhoneNumber'].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)
df['PhoneNumber'].head(50)

## Phone number has been converted into desired format
```

```python
## REMOVING UNWANTED COLUMN ADDRESS

df.drop(columns='Address', inplace=True)
df
```

```python
## REARRANGING COLUMNS 

df = df.reindex(columns=['CustomerID','Name', 'PhoneNumber', 'StreetAddress', 'City/Locality', 'State', 'BookTitle', 'PurchaseDate', 'PurchaseAmount'])
```
